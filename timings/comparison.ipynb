$\textbf{Compare Different Methods - Generate Plots}$

````python
# Import required libraries.
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
````

$\textbf{1. Compare Serial Versions}$

+ ```Original```

+ ```NumPy Vectorised```

+ ```Serial Numba```

+ ```Serial Cython```

````python
# Read in the runtimes for the original version.
original = pd.read_csv("original/runtimes.csv")
````

````python
# Read in the runtimes for the numpy vectorised version.
numpy = pd.read_csv("numpy/runtimes.csv")
````

````python
# Read in the runtimes for the serial numba version.
s_numba = pd.read_csv("serial_numba/runtimes.csv")
````

````python
# Read in the runtimes for the serial cython version
s_cython = pd.read_csv("serial_cython/runtimes.csv")
````

````python
# Plot the runtimes against each other.
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(original["Lattice-Size"], original["Runtime-Average"], "o--", label="Original")
ax.plot(numpy["Lattice-Size"], numpy["Runtime-Average"], "o--", label="NumPy Vectorised")
ax.plot(s_numba["Lattice-Size"], s_numba["Runtime-Average"], "o--", label="Serial Numba")
ax.plot(s_cython["Lattice-Size"], s_cython["Runtime-Average"], "o--", label="Serial Cython")

ax.grid()
ax.legend()

ax.set_title("Lebwohl-Lasher Serial Code - BC4 Runtimes")
ax.set_xlabel("Lattice Size")
ax.set_ylabel("Runtime (s)")

plt.show()
````

$\textbf{5. Compare Runtimes Of Specific Functions}$

+ $\text{Lattice Size: 100}$
+ $\text{Reduced Temperature: 0.5}$

- $\text{Number Of Function Runs: 1000}$

$\textbf{5.1 Compare Function:}$ ```cell_energy```

````python
# Store the runtimes.
cell_energy_versions = ["Original", "Serial Numba", "Serial Cython", "Parallel Numba", "Parallel Cython"]
cell_energy_runtimes = np.array([1.0354345025916701e-05, 1.38431597588351e-06, 1.915916987854871e-06, 9.60865012530121e-07, 1.6554280191485304e-06])

# Calculate the speed-up factors.
cell_energy_speedup = cell_energy_runtimes[0] / cell_energy_runtimes
cell_energy_speedup = np.round(cell_energy_speedup, 2)

# Print the speed-up factors.
print(cell_energy_speedup)
````

````python
# Plot the runtimes of the different versions.
# Exclude the runtime of the original script.
fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(cell_energy_versions[1:], cell_energy_runtimes[1:], color="royalblue")

for i, runtime in enumerate(cell_energy_runtimes[1:]):
    ax.text(i, (runtime + (runtime * 0.01)), f"x{cell_energy_speedup[1:][i]}", ha="center")

ax.set_title("Cell-Energy Function Local Runtimes (Threads = 4)")
ax.set_xlabel("Function Version")
ax.set_ylabel("Runtimes (s)")

plt.show()
````
